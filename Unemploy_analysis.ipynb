```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
#df = pd.read_csv('Unemployment in India.csv')
df = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
sns.pairplot(df, hue='Region')
```

```python
plt.figure(figsize=(13,9))
sns.heatmap(df.corr())
```

```python
df.columns= ['State', 'Date', 'Frequency', 
             'Estimated Unemployment Rate','Estimated Employed',  
             'Estimated Labour Participation Rate','Area','longitude', 'latitude' ]
plt.figure(figsize=(14,9))
plt.title("Indian Employment")
sns.histplot(x="Estimated Employed",hue = 'State',data= df)
plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate",hue = 'Area',data= df)
plt.show()
```

```python
style = sns.axes_style("whitegrid")
style["axes.grid"] = False
sns.set_style("whitegrid", style)
```

```python
figure = plt.figure(figsize=(14, 8))
axe = figure.gca()
df.plot(x="Date", y="Estimated Unemployment Rate", ax=axe, legend=False)
df.plot(x="Date", y="Estimated Unemployment Rate", ax=axe, legend=False)
axe.set_ylabel("% Unemployment")
axe.set_title("Unemployment Rate (Jan 2020 - April 2020)")
```
